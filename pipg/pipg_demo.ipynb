```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
### creating the gird
l = 3
N = 100
x = np.linspace(-l,l,N)
y = np.linspace(-l,l,N)
xx, yy = np.meshgrid(x,y)

h = 2 * x + 2

dummy = 0*x + 5

f = 4.0*xx*xx + 1.0*yy*yy

plt.figure(figsize=(10,10))
plt.contour(xx,yy,f)
plt.scatter(0, 1.5, s=4000)
plt.plot(x,h,'k',label='Equality Constraint')
plt.plot(x,dummy,label='Inequality Constraint')
plt.legend()

plt.xlim((-l,l))
plt.ylim((-l,l))
plt.show()
```
